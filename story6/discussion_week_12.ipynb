I was really inspired by the visualization animations from the linked TED Talk on YouTube. I felt they really communicated clearly a story by simply displaying the changes in the data overtime. Rather than engage in a full critique of that video though, I would like to propose the following experiment, inspired by the visualizations used in the aforementioned video.

An experiment examining the national and global distribution of private wealth over the last 70 years:
Null Hypothesis: There has bene no statistically significant change in the distribution of private wealth over the last 70 years
Hypothesis: There has been a statistically significant change in the distribution of private wealth over the last 70 years.

Data would be collected from Censuses, the OECD, the UN, and similar governmental and non-governmental sources. Analysis performed would be the change in the shape of the distribution over time (including a statistical comparison, the KS test) and the change in inequality metrics overtime (primarily the Gini coefficient).

For visualizations, both of the above distributions would be mapped for each year and then animated to show their changes (or lack of changes) over time. Additional potential visuals might be plotting the percent of wealth held by the percent of population as a curve and its changes overtime. Another on might be a nested proportional area chart or bubble chart that communicates the proportion of the wealth and the number of people who hold it over time, for example how much wealth the top 1% control vs the other 99% over time, as an animation. In all, there are a lot of opportunities to create interesting visuals that communicate this story.
