"The whole is greater than the sum of the parts" is a common expression for the meaning of a Gestalt. Multiple figures in in a single Data Visualization with appropriate labeling and text implement this concept. For your post in this forum explain when and how you will use multiple figures to create high quality data visualizations, and comment on the threads posted by others..



A core principle of data visualization is the ink ratio. This is the notion that it is important to "measure the proportion of ink used to display actual data compared to the ink used in a graphic." This principle encourages the removal of any non-essential elements, to pare down a visualization to its critical and necessary elements.

This principle continues to apply when creating visualizations with multiple figures. The multiple figures must be necessary to communicate the information. While it can often be seductive to layer in fun graphical elements, it is paramount that they be necessary to communicate the message within the larger story.

That is an additional crux of the multiple elements to a graphic, the overall story. How do the multiple elements both convey their own message in conjunction? The juxtaposition of two or more visualizations should each tell their own bit of information, but it is their placement together that should enhance their message they communicate and advance the story being told.

It is this co-placement that, when well executed, creates Gestalt, that makes the co-visualizations more than the sum of the parts. While it would be impossible to state exactly how to do this everytime, there are some guidelines that can be followed beyond the common principles of good data visualizations:

- The elements should explain or enhance each other: Perhaps one element explains a detail or provides insight about the previous element.
- Anticipation: Additional elements should ask and answer questions created by initial elements.
- Elements should have graphical continuity: Color use, proximity, contrast, and similarity should all be contiguous and logically related throughout the multiple elements.
- Labelling: Should be minimal and ideally be universal to all visualization elements.
- Contrast: A smaller version of a bar chart to explain a bar chart is boring. Select visual elements that contrast and heighten the visual explanation of the data, rather than repeat it.
