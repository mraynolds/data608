"Context", a word that is frequently used (ed.G. "You took my statement out of context.") but vaguely understood semantically. For this weeks discussion look at https://www.freethesaurus.com/context or another source of your choosing to understand the meaning of context and then post a thread on the discussion forum indicating how you would modify the Story that you are submitting this week for each of the following audiences.

A high-school class

An Op-Ed piece in the NY Times

An presentation to the local Business Roundtable

A Stats Professional Society of Financial Advisors


Context is everything. Hearing an exclamation like "Don't Shoot!" through a closed door, upon opening the door could reveal a person pleading to not be shot with a gun or a person not wanting their photo taken. Only inside the context is the severity of the statement revealed.

Context is the environment and knowledge that gives either explict or implicit structure into how something will be understood. Often times context is implied, audiences are familiar enough with the topic, the speaker, or the surroundings to understand the context. As often, context must be provided by the communicator so that their message is understood as they intended. We live in a global society where cultures and experience may lead to very different understandings of a message. It is vital to consider the audience and incorporate the context into a message based on how it will be understood.

All this means: tailoring the message. In order to tailor a story about US gun laws and gun related deaths for specific groups, the most impactful message will be one that considers their specific relationships to the topic.

For high school students, there are several additional factors to consider and incorporate. Incorporating shootings in schools as an element to the message, information about the specific state the students are in will help center the message for them. Ensuring that there is context on how the laws work in relation to them at their age and location will also help give them a framework for understanding the situation.

This will be different from how the message is framed for an Op-Ed in the NY Times. The New York Times reaches nearly 12 million subscribers with a median age of 43 who are primarily center-left politically, and are, assumedly, relatively well-informed on global affairs in a way that most high schoolers are not. Giving global context will be good for this audience, an awareness of economic impacts will also be useful, and providing generalized national overview is a good start, but then also drilling down to specific very human stories will ground the editorial and provide meaningful anecdotes that the reader can discuss in the rest of their lives.

The best message will differ still for a local Business Roundtable. Here, specific local and economic impact will be a useful anchor to the message. How do firearms affect the local business community? What are the local laws? How are business gun owners impacted by local laws and gun violence? This is more focused than an international Op-Ed.

Even more specific would be a presentation for a statistics professional society of financial advisors. Here, legal context will be important but even more detailed statistics on financial impacts on a national and global level will be useful. Rather than specific local impact, the specific economic impact is valuable. Additionally, there is the opportunity to be more technically communicative by using more statistics and statistical language to illustrate the point.

In all, setting context is what is going to make a message understood but also what will make it impactful. By giving a framework for the message to be understood, there is also the opportunity for making sure it is felt.
