In our professional endeavors we are frequently asked for an estimate> - "How long will something take to complete; and, how much will the effort cost either in FTE hours or $$$."

That question is fundamentally ill-posed as it presupposes an ability to predict the future and make a statement like, "It will take six weeks and cost 3 FTE months of effort."

The question that should be asked is, "What is the probability that this task can be accomplished within six weeks and cost no more than 3 FTE months of effort"

Presented with the challenge of explaining the probability that a task can be completed within six weeks and cost no more than 3FTE months, and less than $$$ to a non-mathematician, I would either start or end by dragging out the old, trusty: "Iron Triangle."



For those unfamiliar, the "Iron Triangle" is a way of considering the constraints on a project. As the saying goes, when deciding if a project can be good, fast, or cheap, you can only ever pick two. For example, if a project is good and fast, it won't be cheap, and to make it cheap, it will either not be good, or not be fast.

A custom triangle chart of that highlights each range for the project: cost from cheap to expensive, duration from slow to fast, and what a quality standard is from good to poor will help frame the problem conceptually and give an udnerstanding of the range of possible outcomes.

Another possible visualization is tracking previous similar projects, and summarizing their estimated and actual timelines, costs, and success rates.  This could be a slope chart with indications for estimated cost and duration compared to actual cost and duration.

Followed with a line chart with an x-axis for duration of project and two lines, one with a labor vs duration and one with cost vs duration with the intersection at 6 weeks and 3FTE months of effort as a way of depicting the interaction of labor and cost and what the ramifications of being shorter or longer on the project would be. Finally, adding an error bar, that is tuned to the probability of finishing within the parameters to the lines on this chart, would indicate what the lower and upper limits could be under ideal or adverse conditions. If the error bars cover the best possible outcome and the worst possible outcome, their overlap will create the region of possible outcomes for the project and help dictate whether the output is appropriate for the goals.
