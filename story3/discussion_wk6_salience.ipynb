We have all seen the data visualization that has every bell and whistle and color, as though the designer believed the more information they provided, the more effective their visualization would be. With these visualizations though, it is never clear what are the crucial facts being communicated. When all the details are highlighted, there is no way to know which information is the most salient.

In short saliency is about focus. In data visualization this is about using the physical attributes of a visualization to draw attention to crucial facts. Much like focus, this should be done as much by subtraction as by addition. In order to achieve saliency a strategy of paring a visualization down to the bare minimum and then building it back up, highlighting the facts that tell a story. A potential foundation for this strategy could include working through a list of tasks listed below:
1. Understand the data and the story you want to tell.
2. Select an appropriate visualization.
3. Strip away all excess notation from the graphic: color, labels, axes, tick marks, line type, text, etc.
4. Ask the questions: What is the story? What is the purpose of this visualization? What part of this graphic communicates the details of the purpose?
5. Selectively add in physical attributes that answer the questions in number 4.
6. Complete the visualization. Put it away, and look at it again after a short time. Where do you focus? Does the focal pints of the visualization provide the details needed to understand the purpose of the visualization and its role in the story.

This strategy will provide a foundational approach to saliency.
