For this week's discussion post a High Quality Data Visual Template for both a Project Score Card and a Burn Down Chart. You may source template from any public source (with attribution) but should be certain to modify them, if necessary, to meet the Quality Features that we have previously discussed.

The Balanced Scorecard

The task of a template for a balanced scorecard is an interesting one as the balanced scorecard is largely a blending of narrative goals and targets and only a limited expression of data. As a tool it is a more holistic approach for understanding and tracking the motivations and potential impacts in the short and long term for a project and an organization.

The following chart, from Lucid charts(https://www.lucidchart.com/blog/how-to-create-a-balanced-scorecard), boils down this relatively complex expression to a simple grid providing a simple and easily digested summary of the balanced scorecard. Improvement of this template may be possible by highlighting or bolding key elements. Additionally, the color choice as it is purely aesthetic and does not convey meaningful information, marking a place for improvement.

The Burn-Down Chart

As the burn-down chart is useful for visualizing the progress of a project or sprint, it provides a quick and informative sense of a team's progress and in project debriefs as sense of how the team performed. In a practical sense these can be produced rapidly in some common project management software services, such as Jira, which is the source for the image below.

As with the scorecard, the colors do not provide meaning. As we have discussed color is part of the semantic understanding of a high quality visualization.
