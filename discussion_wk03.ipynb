In her lecture "The Sentence is a Lonely Place", Gary Lutz articulates the oft trod concept of writing as rewriting, emphasizing that a writer's economy of words and word selection are simultaneously the writing and the impact of the writing. In much the same bi-dimensional way, the Data Visualization Quality Features, Tufte's Principles, and Cavaller's Principles for Data Visualizations are both a writing and rewriting of each other and a statement about economy of design and the proportional impact of the design.

Fundamental to all three are the primacy of the data and how it is represented. While each set of principles has at least seven parts, they each contain some redundancy or restatement about how the data should be treated and how the data should be perceived. Indeed, nearly all principles can fit directly into these two interconnected categorical questions: How will the data be displayed? and How should the data be perceived?

These two questions are what each set of principles attempts to answer, but in fact, they essentially have the same answer: Honesty. The data visualization should be honest. It should tell the truth of the data in a way that the raw data cannot. If a table is an effective communicator of the data there is no purpose to the data visualization. However, a table rarely is effective especially when a dataset is large or complex. In these instances a visualization has the ability to communicate the honest story of the data more effectively than the raw data.

When it comes to how the data will be displayed, each set of principles attempts to ensure that the data is displayed honestly. The three sets of principles all approach this concept in slightly different ways. The Data Visualization Quality Features do so in the following principles:

- Fidelity: True representation of data, avoiding omissions or additions
- Simplicity: Keep it simple, avoiding unnecessary embellishments
- Utility: Meet viewer's needs, excluding extraneous information
- Saliency: Focus on crucial facts and conclusions, supported by relevant details

Each of these addresses representing the data truthfully, without excess or embellishment, using only what is essential. These principles are echoed in Tufte's principles:

- Consists of complex ideas communicated with clarity, precision, and efficiency
- Requires telling the truth about the data
- Show data variation, not design variation
- Graphics must not quote data out of context
- Show the data

Cavaller also considers and emphasizes honesty in how the data will be displayed:

- Keeps scientific rigor, order, and method in the quantitative procedures
- Graphical representation has a practical utility and added value over the other communicative forms facilitating their comprehension
- Achieves the communication goals by the optimal means of communication with maximum benefits and minimal use of resources

Each of these principles has overlap in honesty: displaying the data effectively and without excess. The other half of the data visualization is an understanding of how the data will be believed, or stated another way: "How is the honestly displayed data being understood?". As a concept this is very related to an understanding of context and has already been touched on by the principles in their unanimity regarding effectively displaying the data without embellishment or excess. However, this group of principles is deserving of their own overarching category because they either directly or indirectly consider the viewer of the data visualization. Whereas displaying the data honestly is the first principle, understanding how it will be perceived is the other half of this concept. Indeed, even if data is displayed honestly and sparingly, if it will be read otherwise or not understood, is the visualization honest or effective? It is not.

With this in mind, each of the three sets of principles addresses how the visualization will be perceived. For the Data Visualization Quality Features all the following principles address perception:

- Utility: Meet viewer's needs, excluding extraneous information
- Efficacy: Achieve the intended purpose within the specified context
- Uniformity: Maintain a consistent presentation themes
- Amity: Prioritize user-friendliness for minimal mental effort

For Tufte the following principles all consider perception:
- Erase non-data-ink/Erase redundant data-ink
- Use words numbers and graphic elements together in concert
- Display an accessible complexity of detail
- Have a narrative quality: a story to tell about the data

For Cavaller the following principles all consider perception:
- Graphical representation has a practical utility and added value over the other communicative forms facilitating their comprehension
- Able to package data message and graphic representation in a singular configuration that promotes the understanding of a meaningful communication
- Taking into account the professional, social, and cultural context and target; it produces intended communicative result by a suitable design
- Achieves the communication goals by the optimal means of communication with maximum benefits and minimal use of resources
- Receives a positive assessment from the user in terms of usability and of  other factors related.

In truth, amongst the two categories presented here and the individual groups of principles themselves, there is an extraordinary amount of overlap. Portions of each could be separately assigned to the two main conceptual categories. This only further highlights the extent to which the goal of all the principles is the same: honesty. The honest communication of the data; everything else serves this principle.

Where the three sets differ is largely a semantic discussion where they differ in type but not kind. However, each stresses different themes, giving each grouping a different central focus. For Data Visualization Quality Features there is a stress on spareness with lines such as: "avoiding omissions or additions", avoiding unnecessary embellishments", "excluding unnecessary embellishments", "focus on crucial facts and conclusions", and "minimal mental effort". This shows a driving focus away from excess and towards only using what is truly necessary.

This concept of spareness appears in Tufte's principles as well: "precision, and efficiency", "the shortest time with the least ink in the smallest space", "erase non-data-ink", and "eliminate chart-junk". However, it is secondary in Tufte's principles. Tufte stresses two concepts with greater force: technical cohesion and storytelling. Concepts such as: "Reflect a balance, a proportion, a sense of relevant scale" and "Use words, numbers and graphic elements together in concert" highlight the principles' focus on technical cohesion. However, Tufte does not sacrifice storytelling at the altar of spareness or technical cohesion, stating: "Have a narrative quality: a story to tell about the data" and "Data visualization are paragraphs about data and should be treated as such".

Of the three sets of principles, Cavaller's do the most to encode evaluation, and thus the concept of perception. Stressing "adaption to the medium", "added value over the other communicative forms", "Achieves the communication goals by the optimal means of communication" results in a grouping of principles that are particularly focused towards context, comprehension, and perception.

To state that any of the sets of principles are NOT focused on context, comprehension, and perception would be inaccurate; they are all concerned with these principles, but all of these principles serve the core idea of honesty, of communicating the data in an honest way. To be honest the communication must be clear, well selected and edited, appropriate for the data type, and without embellishment. (Indeed, what is embellishment if not dishonest or inaccurate?).

Honesty is very much how the three sets of principles end up saying the same things and even repeating themselves because they are working toward that core principle. Perhaps this is a place where the constraints of the English language make communicating a concept difficult, which is further evidence on the value of a well constructed, and honest, data visualization.
