Week 11: Smoothing
Starts Apr 13
For this week your post should identify a use case where you would use data smoothing in creating your data visualization and one where it would not be appropriate because the short term (higher frequency) fluctuations are the salient information for your story.

In recent years I have worked growing plants in greenhouses. Two metrics that we monitor regularly are the temperature and light. Both have several different ways of being measured and at different times and for different purposes it is necessary to display them either smoothed or in their raw state.

The most frequently used temperature metric, and the one I most often graph is the ADT or average daily temperature. This is an average of the night and day temperatures that gives a sense of the average temperatures the plant is experiencing over its lifetime. However, in an environment where there are very hot days and very cold nights, the ADT does not tell the full story, and charting the actual high and low temperatures each day is necessary.

Similarly, one of the most useful measurements is the Daily Light Integral or DLI, which is the amount of photosynthetically active radiation in a 24-hour period. This is often the most useful measurement when thinking about the growth of a plant over its lifecycle. However, it is also important to chart light as PPFD which is the photosynthetic photon flux density which is the instantaneous measure of photons actually hitting the plant in a given moment. A PPFD that is too high may cause damage and stress to a plant. The solution is often to shade the plant during the highest solar radiation times of the day, but you would not necessarily know this was occurring if you only consider a chart of the DLI.
